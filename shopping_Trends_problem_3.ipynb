**3. Advanced Data Aggregation and Plotting:** Using pandas, apply multiple aggregation functions (mean, sum, count) to a dataset grouped by multiple columns. Visualize the aggregated data in a multi-panel (subplot) bar chart using matplotlib.


```python
import pandas as pd
import matplotlib.pyplot as plt
shopping_data = pd.read_csv('/content/shopping_trends.csv')
```

```python
grouped_data = shopping_data.groupby(['Gender', 'Category'])['Purchase Amount (USD)'].agg(['mean', 'sum', 'count'])
grouped_data.head()
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

grouped_data['mean'].unstack().plot(kind='bar', ax=axes[0], title='Mean')
grouped_data['sum'].unstack().plot(kind='bar', ax=axes[1], title='Sum')
grouped_data['count'].unstack().plot(kind='bar', ax=axes[2], title='Count')

plt.tight_layout()
plt.show()
```
