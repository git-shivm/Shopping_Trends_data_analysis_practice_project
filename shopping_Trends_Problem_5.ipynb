# Problem 5
Sum all the Puchase Amount (USD) by their respective category and plot them on line graph

```python
import pandas as pd
import matplotlib.pyplot as plt
shopping_data = pd.read_csv('/content/shopping_trends.csv')
```

```python
category_sales = shopping_data.groupby('Category')['Purchase Amount (USD)'].sum()
category_sales.plot(kind = 'line')
plt.ylabel('Total Sales')
plt.xlabel('Category')
plt.title('Total Sales by Category')
```
