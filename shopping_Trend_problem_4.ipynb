**4. Cumulative Data and Area Plot:** Use pandas to calculate the cumulative sum of a specific column (e.g., cumulative sales over time). Create an area plot using matplotlib to visualize the cumulative data.


```python
import pandas as pd
import matplotlib.pyplot as plt
shopping_data = pd.read_csv('/content/shopping_trends.csv')
```

```python
shopping_data['cumulative_sum'] = shopping_data['Purchase Amount (USD)'].cumsum()
# cumulativesales.plot(kind = 'area')
plt.figure(figsize=(10, 5))
plt.fill_between(shopping_data.index, shopping_data['cumulative_sum'])
plt.xlabel('Index')
plt.ylabel('Cumulative Sum')
plt.title('Cumulative Sum of Amount')
plt.grid(True)
plt.show()
```
