**Data Grouping and Visualization:** Using pandas, group data by a specific column (e.g., "Category") and compute the mean of another column (e.g., "Average Purchase Amount (USD)").
 Create a bar plot using matplotlib to visualize the grouped data.

```python
import pandas as pd
import matplotlib.pyplot as plt
shopping_data = pd.read_csv('/content/shopping_trends.csv')


```

```python
grouped_category = shopping_data.groupby('Category')['Purchase Amount (USD)'].mean()
grouped_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.title('Average Purchase Amount by Category')
plt.show()
```
